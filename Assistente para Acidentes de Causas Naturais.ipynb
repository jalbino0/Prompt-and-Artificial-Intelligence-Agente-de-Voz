```python
import openai
from openai import OpenAI
from google.colab import userdata

client = OpenAI(api_key=userdata.get("OPENAI_API_KEY"))

orientacoes_desastres = {
    "terremoto": (
        "Durante um terremoto:\n"
        "- Fique longe de janelas e objetos que possam cair.\n"
        "- Proteja a cabeça e o pescoço embaixo de mesas resistentes.\n"
        "- Após o tremor, saia com cuidado e fique longe de prédios danificados."
    ),
    "tsunami": (
        "Em caso de alerta de tsunami:\n"
        "- Vá imediatamente para áreas mais altas.\n"
        "- Afaste-se da costa, rios e estuários.\n"
        "- Não volte até a liberação oficial das autoridades."
    ),
    "vulcão": (
        "Durante uma erupção vulcânica:\n"
        "- Use máscaras ou panos úmidos para cobrir o nariz e a boca.\n"
        "- Fique longe de áreas de fluxo de lava ou cinzas.\n"
        "- Siga as instruções da Defesa Civil e evacue se for orientado."
    ),
    "furacão": (
        "Em caso de furacão:\n"
        "- Feche portas e janelas.\n"
        "- Prepare um abrigo com suprimentos essenciais.\n"
        "- Evite sair até o aviso oficial de segurança."
    ),
    "tornado": (
        "Se um tornado for avistado:\n"
        "- Vá imediatamente para um porão ou cômodo sem janelas.\n"
        "- Proteja a cabeça com almofadas ou colchões.\n"
        "- Evite carros e áreas abertas."
    ),
    "enchente": (
        "Durante enchentes:\n"
        "- Não ande nem dirija em áreas alagadas.\n"
        "- Desligue a energia se a água estiver entrando.\n"
        "- Vá para locais altos e seguros."
    )
}

def resposta_por_usuario(tipo_usuario):
    tipo_usuario = tipo_usuario.lower()

    if tipo_usuario == "vítima":
        return (
            "Sinto muito pelo que está passando. Aqui estão alguns passos imediatos:\n"
            "1. Verifique se você está em segurança física.\n"
            "2. Procure um abrigo próximo (ligue 193 ou acesse defesa-civil.gov.br).\n"
            "3. Tente contatar familiares e amigos.\n"
            "4. Se precisar de atendimento médico, procure o posto de saúde mais próximo.\n"
            "Você não está sozinho. Podemos continuar conversando se quiser."
        )

    elif tipo_usuario == "morador":
        return (
            "Obrigado por buscar informação a tempo. Aqui vai o que você pode fazer:\n"
            "1. Fique atento aos alertas da Defesa Civil.\n"
            "2. Prepare uma mochila de emergência com documentos, remédios e água.\n"
            "3. Se o risco for imediato, evacue a área e procure abrigo seguro.\n"
            "4. Evite áreas alagadas ou encostas.\n"
            "Cuide de você e da sua família. Podemos conversar mais se precisar."
        )

    elif tipo_usuario == "familiar":
        return (
            "Entendo sua preocupação. Para saber informações sobre alguém na área afetada:\n"
            "1. Entre em contato com os órgãos oficiais (Defesa Civil ou Corpo de Bombeiros).\n"
            "2. Verifique listas de resgatados e abrigados nos sites oficiais.\n"
            "3. Use redes sociais e grupos locais, mas confirme com fontes confiáveis.\n"
            "4. Mantenha a calma e não desista.\n"
            "Se quiser, posso te ajudar a simular como fazer esse contato."
        )

    else:
        return "Tipo de usuário não reconhecido. Por favor, escolha entre: vítima, morador ou familiar."

def gerar_prompt(tipo_usuario, pergunta):
    prompt = f"Tipo de usuário: {tipo_usuario}\n"
    prompt += f"Dúvida: {pergunta}\n"

    for desastre, orientacao in orientacoes_desastres.items():
        if desastre in pergunta.lower():
            prompt += f"\nInformações sobre {desastre}:\n{orientacao}\n"
            break

    return prompt

print("Bem-vindo ao Assistente de Emergência!")
tipo = input("Você é vítima, morador ou familiar? ").strip()
pergunta = input("Digite sua dúvida ou situação: ").strip()

resposta_inicial = resposta_por_usuario(tipo)

prompt = gerar_prompt(tipo, pergunta)

resposta_api = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "Você é um assistente de emergência que responde com empatia e clareza."},
        {"role": "user", "content": prompt}
    ],
    temperature=0.7
)

resposta_gerada = resposta_api.choices[0].message.content

print("\n--- RESPOSTA DO ASSISTENTE ---\n")
print(resposta_inicial)
print("\nResposta baseada na sua dúvida:\n")
print(resposta_gerada)


```
